```python
import json
import numpy as np
from common.util import array_to_matrix, location_to_point, print_matrix
```

```python
def read_json(filename):
    with open(filename, 'r') as f:
        return json.load(f)
```

```python
DATA_DIR = './data'
```

```python
data = read_json(f'{DATA_DIR}/problem1_day-1.json')
```

```python
N = 5
```

```python
def parse_data(data, N):
    bikes_by_time = []
    returned_bikes = np.zeros((12*60 + 1, N*N), dtype=int)
    prev_bikes = np.zeros(N*N, dtype=int)
    for t in range(12 * 60):
        bikes = prev_bikes + returned_bikes[t]

        if str(t) in data:
            for request in data[str(t)]:
                taking_loc, returning_loc, duration = request
                returning_t = t + duration

                bikes[taking_loc] -= 1
                if returning_t < len(returned_bikes):
                    returned_bikes[t + duration][returning_loc] += 1
        
        bikes += returned_bikes[t]
        bikes_by_time.append(bikes)
        prev_bikes = bikes
    return bikes_by_time
```

```python
N = 60
data = read_json(f'{DATA_DIR}/problem2_day-1.json')
bikes_by_time = parse_data(data, N)
```

```python
from ipywidgets import interactive

def f(t):
    mat = array_to_matrix(bikes_by_time[t])
    print_matrix(mat)

interactive_plot = interactive(f, t=(0, 12*60))
interactive_plot
```
